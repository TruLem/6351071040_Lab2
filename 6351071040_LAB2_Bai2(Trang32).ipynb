# Thông thường để dễ so sánh hai giá trị độ chính xác (precision) và độ phủ (recall) của những phương pháp với nhau người ta sử dụng một giá trị gọi là F-measure. F-measure là trung bình điều hòa (harmonic mean) của hai giá trị độ chính xác và độ phủ được tính bằng công thức:

Hãy tính giá trị F-measure của các bài tập cơ bản.

### Đề bài đơn giản: Tính F-measure

**Tình huống:**
Giả sử tôi có một mô hình phát hiện email spam. Sau khi thử nghiệm mô hình trên một tập dữ liệu nhỏ gồm 10 email, tôi thu được kết quả như sau:

*   **Email thực sự là spam:** 5 email
*   **Email thực sự không phải spam:** 5 email

**Kết quả dự đoán của mô hình:**
*   **Mô hình dự đoán đúng là spam (True Positives - TP):** 4 email
*   **Mô hình dự đoán sai là spam (False Positives - FP):** 1 email (email này thực sự không phải spam)
*   **Mô hình dự đoán sai là không phải spam (False Negatives - FN):** 1 email (email này thực sự là spam)
*   **Mô hình dự đoán đúng là không phải spam (True Negatives - TN):** 4 email

**Yêu cầu:**
Tính độ chính xác (Precision), độ phủ (Recall) và F-measure của mô hình này.

```python
# Định nghĩa các giá trị từ đề bài
TP = 4  # True Positives
FP = 1  # False Positives
FN = 1  # False Negatives

# Tính Precision (Độ chính xác)
# Precision = TP / (TP + FP)
precision = TP / (TP + FP)

# Tính Recall (Độ phủ)
# Recall = TP / (TP + FN)
recall = TP / (TP + FN)

# Tính F-measure (F1-score), sử dụng công thức trung bình điều hòa
# F-measure = 2 * (Precision * Recall) / (Precision + Recall)
f_measure = 2 * (precision * recall) / (precision + recall)

# In kết quả
print(f"Độ chính xác (Precision): {precision:.2f}")
print(f"Độ phủ (Recall): {recall:.2f}")
print(f"F-measure: {f_measure:.2f}")
```
